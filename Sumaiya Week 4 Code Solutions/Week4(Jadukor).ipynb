```python
import re
from itertools import combinations
from banglakit import lemmatizer as lem
from banglakit.lemmatizer import BengaliLemmatizer

# Initialize the BanglaKit Lemmatizer
lemmatizer = BengaliLemmatizer()

# Sample Bengali stop words (expand this list as needed)
BENGALI_STOP_WORDS = set([
    "এটা", "এবং", "অথবা", "যা", "যদি", "তাহলে", "হয়", "হতে", "নিয়ে", "আমার", "তার", "আপনার"
])

def remove_stop_words(text):
    words = text.split()
    filtered_words = [word for word in words if word not in BENGALI_STOP_WORDS]
    return ' '.join(filtered_words)

def bangla_sentence_tokenize(human_written_text):
    # Step 1: Remove stop words
    filtered_text = remove_stop_words(human_written_text)

    # Step 2: Split the filtered text into words for lemmatization
    words = filtered_text.split()

    # Step 3: Apply lemmatization to each word
    lemmatized_words = [lemmatizer.lemmatize(word, pos='proper_noun') for word in words]

    # Step 4: Join the lemmatized words back into a single string
    lemmatized_text = ' '.join(lemmatized_words)

    # Step 5: Tokenize by Bangla full stop (।)
    sentences = lemmatized_text.split('।')

    # Step 6: Remove empty or whitespace-only sentences and strip leading/trailing spaces
    cleaned_sentences = [sentence.strip() for sentence in sentences if sentence.strip()]

    # Return the cleaned sentences and the total number of sentences
    total_sentences = len(cleaned_sentences)

    return cleaned_sentences, total_sentences

def word_pairs_in_sentence(sentences):
    # To store pairs and their counts for each sentence
    sentence_pairs_list = []
    total_pairs_count = []

    # Loop through each sentence
    for sentence in sentences:
        # Tokenize the sentence into words
        words = sentence.split()

        # Generate all possible pairs of words
        word_pairs = list(combinations(words, 2))  # Generate combinations of 2 words

        # Count the number of pairs
        total_pairs = len(word_pairs)

        # Append the pairs and the count to the respective lists
        sentence_pairs_list.append(word_pairs)
        total_pairs_count.append(total_pairs)

    return sentence_pairs_list, total_pairs_count


# Tokenize and get sentences
sentences, total_sentences = bangla_sentence_tokenize(human_written_text)

# Print the results
print(f"Total Number of Sentences: {total_sentences}")

# Get word pairs and total number of pairs for each sentence
sentence_pairs, total_pairs = word_pairs_in_sentence(sentences)

# Print the results
for i, sentence in enumerate(sentences):
    print(f"Sentence {i+1}: {sentence}")
    print(f"Word Pairs: {sentence_pairs[i]}")
    print(f"Total Number of Pairs: {total_pairs[i]}")
```

```python
def count_pair_frequencies(pairs):
    # Flatten the list of pairs and count occurrences
    flat_pairs = [pair for sublist in pairs for pair in sublist]  # Flatten the list of pairs
    frequency_count = Counter(flat_pairs)  # Count the occurrences of each pair
    return frequency_count

def print_pair_frequencies(frequency_count):
    # Print the frequencies in the specified format
    for index, (pair, count) in enumerate(frequency_count.items(), start=1):
        print(f"{index}. {pair} - {count}")
```
